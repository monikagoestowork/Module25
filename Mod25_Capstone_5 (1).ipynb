```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Define the file path to your dataset
file_path = r"C:\Users\monik\health_lifestyle_classification.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# --- Initial Data Inspection ---
print("First 5 rows of the dataset:")
print(df.head())

print("\nNull value counts before cleaning:")
print(df.isnull().sum())


# --- Data Cleaning and Preparation for Plotting ---

# Select the columns needed for the visualization
plot_columns = ['age', 'bmi', 'blood_pressure', 'calorie_intake', 'stress_level', 'target']
plot_df = df[plot_columns].copy()

# Identify numerical columns for imputation
numerical_cols = plot_df.select_dtypes(include=np.number).columns.tolist()

# Handle Missing Values using Imputation for the selected columns
# We use SimpleImputer to fill NaN values with the mean of their respective columns.
imputer = SimpleImputer(strategy='mean')

# Apply the imputer to the numerical columns of our plotting DataFrame
plot_df[numerical_cols] = imputer.fit_transform(plot_df[numerical_cols])

print("\nNull value counts after cleaning for plotting columns:")
print(plot_df.isnull().sum())


# --- Data Visualization ---

# Create a scatter plot matrix (pair plot) to visualize relationships
# The 'hue' parameter colors the points based on the 'target' column ('healthy' vs 'diseased')
print("\nGenerating scatter plot matrix...")
sns.pairplot(plot_df, hue='target', vars=['age', 'bmi', 'blood_pressure', 'calorie_intake', 'stress_level'])

# Add a title to the overall plot
plt.suptitle('Relationship between Health Factors and Target Outcome', y=1.02)

# Display the plot
plt.show()
```
